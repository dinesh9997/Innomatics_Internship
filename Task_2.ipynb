```python
#2
# here i am taking the user input
n=int(input("enter no of entries: "))
d={}
for i in range(n):
   name=input(f"Enter {i+1} Employee name: ")
   sal=int(input(f"Enter {name} salary: "))
   d[name]=sal
print(d)
t=0
for i in d:
   if d[i]>t:
      t=d[i]
      emp=i
      sa=d[i]
print(f"Name {emp} , Highest salary is {sa}")
```
