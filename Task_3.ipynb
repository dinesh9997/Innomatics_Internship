```python
#1
d= {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

m=max(d.values())
s=""
for i in d:
    if(d[i]==m):
        s=s+f" {i},"
s=s[:-1]
print(f"Top Performers Eligible for Bonus: {s} (Score: {m})")  
```

```python
#2
import string
s="Buy mobile phone   buy phone, , online!!"
for i in s:
    if i in string.punctuation:
        s=s.replace(i,"")
s=s.lower()

l=s.split(" ")

d={}
for i in l:
    if i == "":
        continue
    elif i in d:
        d[i]=d[i]+1
    else:
        d[i]=1
print(d)
```

```python
#3
s=[3, 4, 7, 8, 10, 12, 5]
l=[]
for i,j in enumerate(s):
    if(j%2==0):
        l.append([i,j])

print(l)


```

```python
#4
from collections import Counter
s = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
n=len(s)
l=[]
for i in s:
    l.append(i.split("@")[1])
d=Counter(l)
for i in d:
    print(i,": ",((d[i]/n)*100),"%")    


```
